11. **Discount:** Discount given on the product
12. **Profit:** Profit/loss made on the sale

#### **Multivariate Analysis**
From a **business perspective**, it may be useful to **examine how the variables** explored previously, **affect the sales** and most importantly the **profits/losses** gained/incurred by the WOW Superstore.

```python
# Examining Intercorrelation among Numerical Columns

sns.pairplot(data=data, hue = 'Ship Mode')
```

**Quick Observations**
- **Histogram plots** located **along the diagonal** of the figure above simply **reiterate** the **results of** the **univariate analysis** carried out previously
* A lot of orders are clustered between a profit or loss of 1000 dollars.
* The distribution of the variables is similar for both shipment modes, but the number of orders are more for *Standard Class* shipping.
* We had seen that the maximum profit made was ~6720 dollars, and the above plot shows that the order was shipped under *First Class*.
- Interestingly, **Sales and Profit do not have a direct positive correlation** (i.e, *More sales did not yield more profits* for all cases)
- On the other hand, Discount rate is seen to have some correlation to profit.

```python
# Taking a closer look at the relationship between Profits and Sales

sns.scatterplot(data=data, x = 'Sales', y = 'Profit', hue = 'Discount')
```

**Observations**
- The scatterplot above shows more sales actually did not yield more profits for all cases
- For some transactions **increasing sales** amount actually **resulted to larger** financial **losses**. This is **prevalent for those highly discounted purchases**

```python
# Examining the Effect of Discount Rate on Sales and Profits

f, axs = plt.subplots(1, 2, figsize=(8, 4))

# Plotting Sales versus Discount Rate
sns.lineplot(data=data, x="Discount", y="Sales", ax=axs[0])

#Plotting Profits versus Discount Rate
sns.lineplot(data=data, x="Discount", y="Profit", ax=axs[1])

f.tight_layout()
```

**Observations**
- From the plots above, 40% discounted purchases showed a peak in Sales
- On the other hand, the **increasing discount rates show negative effects on the profits** earned. Items purchased at **discounted rates higher than 20% resulted in** significant financial **losses**.


**This gives us an overall idea of how the superstore is performing in financial terms. Digging a little deeper...**
* Let's check the average profit across the different product categories and different states.
* To do so, we will have to first group the data with respect to a particular variable and then check the average profit.


```python
# checking the average profit across the different states

data.groupby("State")["Profit"].mean().sort_values(ascending=False)
```

```python
# checking the average profit across the different product categories

data.groupby("Category")["Profit"].mean().sort_values(ascending=False)
```

**Observations**
* Orders placed from the state of **Washington** result in **more profit**s on average, while those placed from the state of Illinois result in more losses on average.
* Products under the **Furniture** category **result in more losses** on average, while products under the *Technology* category result in more profits on average.

```python
# Checking the distribution of product categories across the different states

data_cross = pd.crosstab(data["State"], data["Category"])
data_cross
```

```python
# Displaying the Crosstab results as a heatmap

sns.heatmap(data_cross, annot=True, fmt=".0f", cmap="PuBu")
```

* Most orders are for office supplies across all the states, where Texas has the highest number of orders.
* The distribution of *Furniture* and *Technology* products are similar for all states except Illinois.
